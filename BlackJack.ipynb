```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing = True
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                new_list = Card(suit,rank)
                self.deck.append(new_list)
    def __str__(self):
        
        deck_comp = ''
        for Card in self.deck:
            deck_comp += '\n '+ Card.__str__()
        return 'This is your deck: ' + deck_comp
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
#the player basically
class Hand:
    #player starts off wit cards and a value is attachted to thoser cards
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
#take in a card object that is called from card deck then its being append the the self.cards list
    def add_cards(self,card):
        
        self.cards.append(card)

#looks up rank in the values dictonary then add it to the self.value of the hand
        self.value +=values[card.rank]
        
        if card.rank == 'Ace':
            self.aces += 1
    def adjust_for_ace(self):
        
        # IF TOTAL VALUE > 21 AND I STILL AN ACE
        # THAN CHANGE MY ACE TO BE A 1
        while self.value > 21 and self.aces > 0 :
            self.value -= 10 
            self.aces -= 1
        
        
        
```

```python
#adds and subs bets from the players total value
class Chips:
    
    def __init__(self):
        self.total = 100 
        self.bet = 0
    
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("How many chips are you betting? "))
        except:
            print("Provie a integer")
        else:
            if chips.bet > chips.total:
                print(f"You dont have enough chips), you have: {chips.total}")
            else:
                break
```

```python
def hit(deck,hand):
    #grabs one card from deck
    single_card = deck.deal()
    #adds to the hand
    hand.add_cards(single_card)
    #checks for an ace
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    while True:
        x = input('Hit or stand? Enter h or s')
        
        if x[0].lower == 'h':
            hit(deck,hand)
            
        elif x[0].lower() == 's':
            print("Player stands dealers turn")
            playing = False
        
        else:
            print("Enter h or s")
            
        break
```

```python
def show_some(player,dealer):
    print("\nDealer's Hand:")
    print(" <card hidden>")
    print('',dealer.cards[1])  
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    
def show_all(player,dealer):
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print("Player's Hand =",player.value)
```
